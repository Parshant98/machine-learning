```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report
import pandas as pd
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df
```

```python
x=df.drop(['class'],axis=1)
y=df['class']
```

```python
y.value_counts()
```

```python
x.describe()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=None)
```

```python
Dmodel=DecisionTreeClassifier(criterion='entropy',max_depth=3)
```

```python
Dmodel=Dmodel.fit(x_train,y_train)
```

```python
y_pred=Dmodel.predict(x_test)
```

```python
print('accuracy',metrics.accuracy_score(y_pred,y_test))
```

```python
y_predtrain=Dmodel.predict(x_train)
```

```python
print('accuracy',metrics.accuracy_score(y_predtrain,y_train))
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data=StringIO()
export_graphviz(Dmodel,out_file=dot_data,filled=True,rounded=True,special_characters=True,class_names=['tested_negative','tested_positive'])
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('dibetes_png')
Image(graph.create_png())
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_random_forest = RandomForestClassifier(n_estimators=170)
clf_random_forest.fit(x_train, y_train)
y_pred = clf_random_forest.predict(x_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python

from sklearn.svm import SVC
clf_svm = SVC()
clf_svm.fit(x_train, y_train)
y_pred = clf_svm.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_std = scaler.fit_transform(x)
# Split dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x_std, y, test_size=0.3, random_state=1) # 70% training and 30% test
from sklearn.svm import SVC
clf_svm = SVC(C=5)
clf_svm.fit(x_train, y_train)
y_pred = clf_svm.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python

from sklearn.linear_model import LogisticRegression
clf_logistic = LogisticRegression()
clf_logistic = clf_logistic.fit(x_train, y_train)
y_pred = clf_logistic.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
